```python
def rng(m=106, a=26, c=5):
    rng.current = (a*rng.current + c) % m
    return rng.current/m

# setting the seed
rng.current = 36

L = [round(rng(),7) for i in range(500)]

len(set(L))
```

```python
import numpy as np
import matplotlib.pyplot as plt

#make this example reproducible
np.random.seed(1)

#generate array of 200 values that follow uniform distribution with min=5 and max=10
data1 = np.random.uniform(low=0, high=1, size=1000)
np.random.seed(1)
data2 = np.random.uniform(low=2, high=3, size=1000)


#create histogram to visualize distribution of values
plt.hist(data1, bins=30, edgecolor='black')
plt.show()
#create histogram to visualize distribution of values
plt.hist(data2, bins=30, edgecolor='black')
plt.show()
```

```python
''' Statsmodels '''

import statsmodels.api as sm
lags = range(10)
acorr1 = sm.tsa.acf(data1, nlags = len(lags)-1)
acorr2 = sm.tsa.acf(data2, nlags = len(lags)-1)

plt.plot(acorr1)
plt.show()
plt.plot(acorr2)
plt.show()
```
